```python
duplicated_rows = df[df.duplicated()]
duplicated_rows
```

```python
#Removing duplicates
df= df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
median_age = df['Age'].median()
median_age
```

```python
# Impute null values in 'Age' column with median
df['Age'].fillna(median_age, inplace=True)
```

```python
median_family = df['Family size'].median()
median_family
```

```python
# Impute null values in 'Famil Size' column with median
df['Family size'].fillna(median_family, inplace=True)
```

```python
# Dropping rows with null values in the "Pin Code" column
df.dropna(subset=['Pin code'], inplace=True)
```

```python
# Impute null values in 'Feedback' column with "No Feedback"
df['Feedback'].fillna('No Feedback', inplace=True)
```

```python
# Checking if there any null values present
df.isnull().sum()
```

```python
#Dropping unwanted Column: Here the "Unnamed 12" is not required
df.drop(columns=['Unnamed: 12'], inplace=True)
```
