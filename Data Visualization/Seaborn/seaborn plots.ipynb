```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_values = [1, 2, 3, 4, 5]
y_values = [10, 15, 20, 25, 30]
category = ['Low', 'Medium', 'Low', 'Low', 'Medium']
```

## Line Plot

```python
sns.lineplot(x=x_values, y=y_values, hue=category, marker='*', palette='Set1')
plt.title('Lineplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Category')
plt.show()
```

## Barplot

```python
sns.barplot(x=x_values, y=y_values, hue=category, palette='Set2')
plt.title('Barplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Category')
plt.show()
```

## Violin Plot

```python
sns.violinplot(x=category, y=y_values, hue=category, palette='Set3')
plt.title('Violinplot')
plt.xlabel('Category')
plt.ylabel('Y-axis')
plt.show()
```

## Boxplot

```python
sns.boxplot(x=category, y=y_values, palette='Pastel1')
plt.title('Boxplot')
plt.xlabel('Category')
plt.ylabel('Y-axis')
plt.show()
```

## Distribution plot

```python
sns.histplot(y_values, kde=True, color='skyblue')
plt.title('Distribution')
plt.xlabel('Y-axis')
plt.ylabel('Density')
plt.show()
```

## Line plot

```python
sns.lineplot(x= 'total_bill', y='tip', hue='sex', marker='*', palette='Set1', data = df)
plt.title('Line Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.show()
```

## Barplot

```python
sns.barplot(data=df, x="day", y="total_bill", hue="sex", palette='Set2')
plt.title('Total Bill by Day with Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Gender')
plt.show()
```

## Violin Plot

```python
sns.violinplot(data=df, x="sex", y="total_bill", palette='Set3')
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.show()
```

## Boxplot

```python
sns.boxplot(data=df, x="sex", y="total_bill", hue="sex", palette='Pastel1', legend=False)
plt.title('Total Bill Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.show()
```

## Distribution Plot

```python
sns.histplot(df["total_bill"], kde=True, color='skyblue')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```
